# 3.8 用 Harness Engineering 和 Loop Engineering 打造可上线的业务能力

## 🚄 前言

你已经学习了如何用 Qwen Code 处理工程任务，也知道 Skill 可以把经验固化下来，评测可以检查结果。下一步要解决的是：一个 Agent 任务怎样从“这一次能跑”，变成“换主题、换要求后也能稳定复用”。

本节课用技术文档流程图做练习。你会让 Qwen Code 先生成一个插图 Skill，再给它补上标准、浏览器截图、验证报告、Reviewer 角色、运行日志和停止条件，让它按证据持续改进。

重点不是画出一张图，而是学习如何把一类 Agent 工作流沉淀成可检查、可复用、可上线的业务能力。


## 🍁 课程目标

学完本节课程后，你将能够：

* 为一个业务任务补齐标准、检查、记录和停止条件
* 让 Qwen Code 生成并改进一个可复用 Skill
* 判断问题应该改作品、改 Skill、补验证器还是补标准
* 解释 Harness Engineering 和 Loop Engineering 的核心作用
* 判断一个 Agent 工作流是否具备上线和持续改进的基础


## 🔎 课程预览

Skill 写出来以后，还要证明它能复用。

本节课用“技术文档流程图生成 Skill”做练习，带你把一次生成任务改造成一条可验证的工作流：先定义标准，再生成样本，用浏览器截图和验证报告发现问题，把可复用经验写回 Skill 或验证器。

核心闭环如下：

```markdown
标准 -> Skill -> 测试样本 -> 截图验证 -> 问题记录 -> 写回规则 -> 新主题验证
```

这条链路跑通后，你得到的不是一张图，而是一套能持续改进的 Agent 工作能力。
